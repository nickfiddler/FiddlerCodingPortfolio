This program will use Random Forest Regressions to predict the number of home runs that an MLB player will hit in a season based off data downloaded from MLB's Statcast

```python
#Reading the file
import pandas as pd

file_name = 'HRFinalAll.csv'
bf = pd.read_csv(file_name)

print(bf)
```

Importing libraries that I will use such as mean squared error, test split, matplotlib, numpy, and the Random Forest Regressor

```python
#Show how they each statistic is correlated to number of home runs

bf_num = bf.apply(pd.to_numeric, errors='coerce')
corr_matrix = bf_num.corr()
hrcorr = corr_matrix["home_run"].sort_values(ascending=False)
hrcorr
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

```python
X1 = bf[['exit_velocity_avg','launch_angle_avg','flyballs_percent','barrel_batted_rate']]
y = bf[['home_run']]

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y, test_size=0.2, random_state=42)
rf_regressor1 = RandomForestRegressor(n_estimators=250, random_state=42)
rf_regressor1.fit(X1_train, y1_train)
y_pred1 = rf_regressor1.predict(X1_test)
```

```python
rmse1 = np.sqrt(mean_squared_error(y1_test, y_pred1))
rmse1
```

```python
X2 = bf[['barrel_batted_rate','hard_hit_percent','exit_velocity_avg','flyballs_percent','solidcontact_percent','launch_angle_avg','pull_percent','whiff_percent','out_zone_percent','z_swing_miss_percent','oz_swing_miss_percent']]

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y, test_size=0.2, random_state=42)
rf_regressor2 = RandomForestRegressor(n_estimators=200, random_state=42)
rf_regressor2.fit(X2_train, y2_train)
y_pred2 = rf_regressor2.predict(X2_test)
```

```python
rmse2 = np.sqrt(mean_squared_error(y2_test, y_pred2))
rmse2
```
