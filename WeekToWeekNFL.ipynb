This program is used as insights to sports gambling. I take tables off of teamrankings.com using beautiful soup, then use pandasql in order to gain information off of the table. As of now, I just have it searching for high ranked offenseses vs low ranked defenses, and offenses and defenses who have been hot (over 3 tds a game) and cold defenses (over 3 tds a game) in the last three games. I update this program week by week to help inform me of where to put anytime touchdown tds. This program showing on github was for Week 10 of the 2024 NFL Season

```python
#Importing packages
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#Websites and creating the data frames
urlOffTds = 'https://www.teamrankings.com/nfl/stat/offensive-touchdowns-per-game'
dfOffTds = pd.DataFrame()
urlDefTds = 'https://www.teamrankings.com/nfl/stat/opponent-touchdowns-per-game'
dfDefTds = pd.DataFrame()
urlRushTds = 'https://www.teamrankings.com/nfl/stat/rushing-touchdowns-per-game'
dfRushTds = pd.DataFrame()
urlPassTds = 'https://www.teamrankings.com/nfl/stat/passing-touchdowns-per-game'
dfPassTds = pd.DataFrame()
urlDefRush = 'https://www.teamrankings.com/nfl/stat/opponent-rushing-touchdowns-per-game'
dfDefRush = pd.DataFrame()
urlDefPass = 'https://www.teamrankings.com/nfl/stat/opponent-passing-touchdowns-per-game'
dfDefPass = pd.DataFrame()

urlList = [urlOffTds,urlDefTds,urlRushTds,urlPassTds,urlDefRush,urlDefPass]
dfList = [dfOffTds,dfDefTds,dfRushTds,dfPassTds,dfDefRush,dfDefPass]
counter = 0
```

```python
#this goes through the list I created from each websites and grabs the table from it
for url in urlList:
  page = requests.get(url)
  soup = BeautifulSoup(page.text,'html')

  table = soup.find_all('table')[0]
  th_list = table.find_all('th')
  titles = [title.text.strip() for title in th_list]

  df = pd.DataFrame(columns = titles)

  column_data = table.find_all('tr')
  for row in column_data[1:]:
    row_data = row.find_all('td')
    ind_row = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = ind_row
  dfList[counter] = df.copy()
  counter = counter+1
```

```python
#For some reason that I do not know, i had to set each table to the spot on
#its list to fill it
dfOffTds = dfList[0]
dfDefTds = dfList[1]
dfRushTds = dfList[2]
dfPassTds = dfList[3]
dfDefRush = dfList[4]
dfDefPass = dfList[5]
```

```python
#Workign with some sql
import pandasql as psql
x = "Philadelphia"
y = "Dallas"
query1 = f'SELECT * FROM dfOffTds WHERE Team = "{x}";'
result1 = psql.sqldf(query1,locals())
print("Offensive Tds")
print(f"{result1}")
query2 = f'SELECT * FROM dfDefTds WHERE Team = "{y}";'
result2 = psql.sqldf(query2,locals())
print("Tds Given Up")
print(f"{result2}")
```

```python
#Away teams and home teams. It is important the same game is in the same position
ateams = ["Cincinnati","Pittsburgh","Philadelphia","NY Giants","Denver","Minnesota","Atlanta","San Francisco","Buffalo","New England","Tennessee","NY Jets","Detroit","Miami"]
hteams = ["Baltimore","Washington","Dallas","Carolina","Kansas City","Jacksonville","New Orleans","Tampa Bay","Indianapolis","Chicago","LA Chargers","Arizona","Houston","LA Rams"]

#Result Lists
goodOffense = []
badDeffense = []
last3Off = []
last3Def = []

#Looping through each game and checking for ranks and last 3 tds
for i in range(0,len(ateams)):
  offQuery = f'Select * FROM dfOffTds WHERE Team = "{ateams[i]}";'
  defQuery = f'Select * FROM dfDefTds WHERE Team = "{hteams[i]}";'
  offQuery2 = f'Select * FROM dfOffTds WHERE Team = "{hteams[i]}";'
  defQuery2 = f'Select * FROM dfDefTds WHERE Team = "{ateams[i]}";'

  offResult = psql.sqldf(offQuery,locals())
  defResult = psql.sqldf(defQuery,locals())
  offResult2 = psql.sqldf(offQuery2,locals())
  defResult2 = psql.sqldf(defQuery2,locals())

  offRank = int(offResult['Rank'][0])
  defRank = int(defResult['Rank'][0])
  offRank2 = int(offResult2['Rank'][0])
  defRank2 = int(defResult2['Rank'][0])

  awayOffLast3 = float(offResult['Last 3'][0])
  homeDefLast3 = float(defResult['Last 3'][0])
  homeOffLast3 = float(offResult2['Last 3'][0])
  awayDefLast3 = float(defResult2['Last 3'][0])

  if offRank <=8 and defRank>=24:
    goodOffense.append(ateams[i])
    badDeffense.append(hteams[i])

  if offRank2 <=8 and defRank2>=24:
    goodOffense.append(hteams[i])
    badDeffense.append(ateams[i])

  if awayOffLast3 >= 3.0 and homeDefLast3 >= 3.0:
    last3Off.append(ateams[i])
    last3Def.append(hteams[i])

  if homeOffLast3 >= 3.0 and awayDefLast3 >= 3.0:
    last3Off.append(hteams[i])
    last3Def.append(ateams[i])

#Printing the results
print(f"Good Offenses: {goodOffense}")
print(f"Bad Defenses: {badDeffense} \n")
print(f"Hot Offenses: {last3Off}")
print(f"Cold Deffenses: {last3Def}")
```
